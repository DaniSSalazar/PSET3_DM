```python
#descarga de los archivos de yellow trips
import os

BASE_URL = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{year}-{month:02d}.parquet"
DOWNLOAD_DIR = "/home/work"

# 2015-2024: todos los meses
# 2025: solo hasta agosto (8)
for year in range(2015, 2026):
    max_month = 8 if year == 2025 else 12
    for month in range(1, max_month + 1):
        url = BASE_URL.format(year=year, month=month)
        local_path = f"{DOWNLOAD_DIR}/yellow_{year}_{month:02d}.parquet"
        
        if os.path.exists(local_path):
            print(f"⏩ Ya existe, omitido: {local_path}")
            continue
        
        print(f"⬇️ Descargando {url}...")
        exit_code = os.system(f"wget -q -O {local_path} {url}")
        
        if exit_code == 0:
            print(f"✅ Guardado en {local_path}")
        else:
            print(f"❌ Error al descargar {url}")

```
